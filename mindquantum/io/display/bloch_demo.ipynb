# Demo for bloch

```python
# bloch label
label = ['x', 'y', ['$\\left|0\\right>$', '$\\left|1\\right>$']]
```

```python
# generate amplitudes
w = np.matrix([[1., 0.]])
h = 1. / np.sqrt(2.) * np.matrix([[1, 1], [1, -1]])
ang = np.pi / 8
rz = np.matrix([[np.exp(-.5 * 1j * ang), 0],
                [0, np.exp(.5 * 1j * ang)]])
amps = []
amps.append(w)
w = w * h
amps.append(w)
for _ in range(6):
    w = w * rz
    amps.append(w)
amps = np.array(amps)
```

```python
# plot
fig = plt.figure(figsize=(20, 10))
for i in range(8):
    ax = fig.add_subplot(2, 4, i + 1, projection='3d')
    amp = amps[i][0]
    draw_single_qubit_bloch(amp, ax=ax, label=label)
    # set title
    title = str(round(amp[0], 2)) + '$\\left|0\\right>$\n'
    title += str(round(amp[1], 2)) + '$\\left|1\\right>$'
    ax.set_title(title)
plt.show()
```
